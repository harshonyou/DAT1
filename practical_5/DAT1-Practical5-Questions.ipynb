**T16i) Write a function moneyWithoutCurrencySymbol that takes a string and returns the string as a float without the currency symbol (if exists)**

```python
#Write your answer here
def moneyWithoutCurrencySymbol(str):
    if str=="NaN":
        return float(0)
    if "$" in str:
        return float(str[1:])
    return float(str)

# print(moneyWithoutCurrencySymbol("$2.3"))
```
