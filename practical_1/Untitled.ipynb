```python
def maximumOccurringCharacter(text):
    # Write your code here
    cha = {}
    for x in text:
        if x in cha:
            cha[x]+=1
        else:
            cha[x] = 1
    cha = dict(sorted(cha.items(), reverse=True,key=lambda item: item[1]))
    return list(cha.keys())[0]

print(maximumOccurringCharacter("hellokitty"))
```
